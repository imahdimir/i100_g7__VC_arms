# Finding Investor IDs for the selected Funds from the 'DealInvestorRelation.csv'

```python
df_deal_inv_rel = pd.read_parquet(file.pb_deal_inv_rel__parq)
df_deal_inv_rel
```

```python
# keep only rows that contains any of the fund Ids we have
```

```python
msk = df_deal_inv_rel[var.investor_fund_id].isin(all_ids)
df_filtered_deal_inv_rel = df_deal_inv_rel[msk]
df_filtered_deal_inv_rel
```

```python
# only keep these columns, I want to add them to the funds spreadsheet later
cols_2_keep = {
        1: var.investor_id,
        2: var.investor_name,
        3: var.investor_fund_id,
        4: var.investor_fund_name,
}

df_filtered_deal_inv_rel = df_filtered_deal_inv_rel[list(cols_2_keep.values())]
df_filtered_deal_inv_rel
```

```python
# Drop duplicates, because we only want unqiute items to match them
df_filtered_deal_inv_rel = df_filtered_deal_inv_rel.drop_duplicates()
df_filtered_deal_inv_rel
```

```python
df_filtered_deal_inv_rel.shape
```

```python
df_filtered_deal_inv_rel.columns
```

```python
# Group by Fund and aggregate Investor info into lists
_df = df_filtered_deal_inv_rel
grouped = _df.groupby([var.investor_fund_id, var.investor_fund_name]).agg(
        {var.investor_id: list, var.investor_name: list}).reset_index()
grouped
```

```python
# Expand lists into separate column
max_investors = grouped[var.investor_id].apply(len).max()
max_investors
```

```python
for i in range(max_investors):
    grouped[f'{var.investor_id}_{i + 1}'] = grouped[var.investor_id].apply(lambda x: x[i] if i < len(x) else None)
    grouped[f'{var.investor_name}_{i + 1}'] = grouped[var.investor_name].apply(lambda x: x[i] if i < len(x) else None)

# Drop original list columns
grouped = grouped.drop(columns = [var.investor_id, var.investor_name])
```

```python
grouped
```

```python
grouped.shape
```

```python
df_selected_funds_2 = pd.merge(df_selected_funds, grouped, left_on = var.fund_id, right_on = var.investor_fund_id,
                               how = 'outer')
df_selected_funds_2
```

```python
df_selected_funds_2.to_excel(file.dm_p030_10__selected_funds__xl, index = False)
```
