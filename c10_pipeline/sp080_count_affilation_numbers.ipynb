# GOAL
- Count affiliation number of VCs and hospitals in the affiliation data

```python
import sys
import pandas as pd
import numpy as np
from pathlib import Path
from collections import Counter

sys.modules.pop('a20_config.a10_config', None)  # force reload
from a20_config.a10_config import dyr, file, const, var

Path.cwd()
```

```python
df00_aff = pd.read_excel(file.dp_sp060_10__all_hospitals_and_VC_affiliations__final__xl)
df00_aff
```

```python
# Count occurrences of each affiliate_id for VCs
df = df00_aff.copy()

msk = df[var.affiliate_type].eq(const.vc)
df = df[msk]

affiliate_counts = df[var.affiliate_id].value_counts()
affiliate_counts
```

```python
# Count how many affiliates appear 1 time, 2 times, 3 times, etc.
frequency_distribution = Counter(affiliate_counts)
frequency_distribution
```

```python
# Convert to a sorted DataFrame for better visualization
freq_df = pd.DataFrame(list(frequency_distribution.items()),
                       columns = ['Count_of_Affiliates', 'Number_of_Occurrences'])
freq_df = freq_df.sort_values('Count_of_Affiliates').reset_index(drop = True)

print("Frequency Distribution of Affiliate IDs:")
print("=" * 50)
print(freq_df.to_string(index = False))
print("\n" + "=" * 50)
print(f"Total unique affiliate IDs: {len(affiliate_counts)}")
print(f"Maximum count of affiliates: {affiliate_counts.max()}")
print(f"Total records: {len(df)}")

# Optional: Create a more detailed summary
print("\nDetailed Breakdown:")
print("-" * 50)
for occurrences in sorted(frequency_distribution.keys()):
    count = frequency_distribution[occurrences]
    percentage = (count / len(affiliate_counts)) * 100
    print(f"{count} affiliate(s) appeared {occurrences} time(s) ({percentage:.1f}%)")
```
