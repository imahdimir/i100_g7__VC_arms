# ReFiltering Hospital-VC affiliations
now only keep those VC that has deal data in this clean deal dataset

```python
df = df_deal_vc_5.copy()

df01_deal_inv_rel = pd.read_parquet(file.pb_deal_inv_rel__parq)
df_1 = df01_deal_inv_rel.copy()

msk = df_1[var.deal_id].isin(df[var.deal_id])

df00_vc_inv = df_1[msk]
df00_vc_inv
# so the final set of VCs should be a subset of this set of investor ids (this list consist of VCs and other types of investors)
```

```python
# so now I need to keep only those affiliations that are in this set of VCs
df_aff = pd.read_excel(file.dp_sp060_10__all_hospitals_and_VC_affiliations__final__xl)
df_aff
```

```python
# drop rows where affiliate type is vc but not in this list of investor ids
# drop rows where enitity type is vc and enitity id not in df00_vc_inv[investor_id]
# make sure all pair number/IDs are exactly twice in the final set
inv_ids = df00_vc_inv[var.investor_id].unique()
len(inv_ids)
```

```python
msk_a = df_aff.apply(
        lambda row: not (row[var.affiliate_type] == const.vc and row[var.affiliate_id] not in inv_ids),
        axis = 1)

msk_b = df_aff.apply(
        lambda row: not (row[var.entity_type] == const.vc and row[var.entity_type] not in inv_ids),
        axis = 1)

msk = msk_a | msk_b

df05_aff = df_aff[msk]
df05_aff
```

```python
# nothing is drop here, the number of rows in df05_aff is the same as df_aff
len(df05_aff), len(df_aff)
#   so there is no need to re-filter the affiliations data again (I wanted to keep only those VCs that have deals in the cleaned deal dataset but all of them already have deals in the cleaned deal dataset)
```
